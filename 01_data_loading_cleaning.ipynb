```python
def schema_report(df):
    return pd.DataFrame({
        "dtype": df.dtypes,
        "missing_count": df.isna().sum(),
        "missing_pct": (df.isna().mean() * 100).round(2)
    })
```

```python
def parse_dates(df):
    cols = [c for c in df.columns if "date" in c.lower() or "time" in c.lower()]
    for c in cols:
        df[c] = pd.to_datetime(df[c], errors="coerce")
    return df, cols
```

```python
enrol_df, enrol_date_cols = parse_dates(enrol_df)
demo_df, demo_date_cols = parse_dates(demo_df)
bio_df, bio_date_cols = parse_dates(bio_df)

enrol_date_cols, demo_date_cols, bio_date_cols
```

# Geographic Standardization

We normalize geography without inventing categories.

Rules applied:

State / district → uppercase trimmed strings

PIN → numeric coercion only

```python
def standardize_geography(df):
    for c in df.columns:
        if "state" in c.lower() or "district" in c.lower():
            df[c] = df[c].astype(str).str.strip().str.upper().replace("NAN", np.nan)
        if "pin" in c.lower():
            df[c] = pd.to_numeric(df[c], errors="coerce")
    return df
```

```python
enrol_df = standardize_geography(enrol_df)
demo_df = standardize_geography(demo_df)
bio_df = standardize_geography(bio_df)
```

# PIN Code Integrity Check

```python
def pin_length_report(df):
    pins = [c for c in df.columns if "pin" in c.lower()]
    return {
        c: round((df[c].dropna().astype(str).str.len() != 6).mean() * 100, 2)
        for c in pins
    }
```

```python
pin_length_report(enrol_df)
```

```python
pin_length_report(demo_df)
```

```python
pin_length_report(bio_df)
```

# Numeric Column Coercion

We convert volume/count fields explicitly.

```python
def clean_numeric(df):
    for c in df.columns:
        if any(k in c.lower() for k in ["count", "number", "total"]):
            df[c] = pd.to_numeric(df[c], errors="coerce")
    return df
```

```python
enrol_df = clean_numeric(enrol_df)
demo_df = clean_numeric(demo_df)
bio_df = clean_numeric(bio_df)
```

```python
def missing_summary(df):
    return (
        df.isna()
        .sum()
        .to_frame("missing_count")
        .assign(missing_pct=lambda x: (x["missing_count"] / len(df) * 100).round(2))
        .sort_values("missing_pct", ascending=False)
    )
```

# Cell — Generate Summaries

```python
enrol_missing = missing_summary(enrol_df)
demo_missing = missing_summary(demo_df)
bio_missing = missing_summary(bio_df)

enrol_missing.head(10)
```
